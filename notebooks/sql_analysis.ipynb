```python
# SQLiteに接続
import sqlite3
import pandas as pd

db_path = "../data/imdb_movies.db"
conn = sqlite3.connect(db_path)

# SQLクエリの実行
query = """
SELECT Series_Title, Released_Year, Genre, IMDB_Rating
FROM movies
WHERE IMDB_Rating > 8.0
ORDER BY IMDB_Rating DESC
LIMIT 10;
"""


df = pd.read_sql(query, conn)
conn.close()

# 結果表示
df
```
